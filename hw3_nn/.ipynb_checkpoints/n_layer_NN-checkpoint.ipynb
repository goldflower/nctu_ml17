```python
img_width = 30
img_height = 30
file_path = 'Data_Train'
class_names = ['Class1', 'Class2', 'Class3']
```

```python
def get_data(img_width = 30, img_height = 30,
             file_path = 'Data_Train', class_names = None,
             validation_ratio = 0.2, nb_samples = [1000,1000,1000]):
    if class_names is None: # testing, no answer
        img_path = [str(a)+'.bmp' for a in range(1, 601)]
        print(img_path)
        df = pd.DataFrame(index = range(len(img_path)), columns = range(img_height*img_width))
        for file, i in zip(img_path, range(len(img_path))):
            img = misc.imread(file_path + '/' + file).flatten()
            df.iloc[i] = img
        return df
    
    train_df = pd.DataFrame(columns = range(img_height*img_width))
    train_df['class'] = 0
    train_df
    for i in range(len(class_names)):
        path = file_path + '/' + class_names[i] + '/'
        print(path)
        class_img_files = os.listdir(path)[:nb_samples[i]]
        
        df = pd.DataFrame(index = range(len(class_img_files)), columns = range(img_height*img_width))
        for file, j in zip(class_img_files, range(len(class_img_files))):
            img_path = path + file
            img = misc.imread(img_path).flatten()
            df.iloc[j] = img
        df['class'] = class_names[i]
        train_df = train_df.append(df)

    np.random.seed(0)
    train_df = train_df.sample(frac=1)
    train_df, val_df = train_df.iloc[:int(len(train_df)*(1-validation_ratio))], train_df.iloc[int(len(train_df)*(1-validation_ratio)):]
    
    train_class = train_df['class']
    train_class.index = range(len(train_class))
    train_df = train_df.drop('class', axis=1)
    train_df.index = range(len(train_df))
    val_class = val_df['class']
    val_class.index = range(len(val_class))
    val_df = val_df.drop('class', axis=1)
    val_df.index = range(len(val_df))
    
    return train_df, train_class, val_df, val_class
```

```python
#### Libraries
# Standard library
import pickle
import gzip

# Third-party libraries
import numpy as np

def load_data():
    f = open('data/mnist.pkl', 'rb')
    
    training_data, validation_data, test_data = pickle.load(f, encoding='iso-8859-1')
    f.close()
    return (training_data, validation_data, test_data)

def load_data_wrapper():
    tr_d, va_d, te_d = load_data()
    training_inputs = [np.reshape(x, (784, 1)) for x in tr_d[0]]
    training_results = [vectorized_result(y) for y in tr_d[1]]
    training_data = zip(training_inputs, training_results)
    validation_inputs = [np.reshape(x, (784, 1)) for x in va_d[0]]
    validation_data = zip(validation_inputs, va_d[1])
    test_inputs = [np.reshape(x, (784, 1)) for x in te_d[0]]
    test_data = zip(test_inputs, te_d[1])
    return (list(training_data), list(validation_data), list(test_data))

def vectorized_result(j):
    e = np.zeros((10, 1))
    e[j] = 1.0
    return e
```

```python
import numpy as np
import random
class NeuronNetwork():
    
    # auto=True: you don't have to create output layer
    #            we'll add it automatically
    # auto=False: the last layer you added is your output layer
    def __init__(self, auto=True):
        np.random.seed(0)
        # set input layer as layer 0
        self.num_layers = 1
        # fit any data yet?
        self.fitted = False
        self.weights = {}
        self.biases = {}
        self.num_neurons = {}
        self.auto = auto
        self.class_names = []
        self.activation_functions = {'sigmoid':self.sigmoid, 'relu':self.relu, 'tanh':self.tanh}
        self.f = None
        self.seesee = {}
        self.int_to_class_name = {}
        self.weight_adjustments = None
        self.count = 0

    def add_layer(self, num_neurons, initialization='normal', input_size=None):
        # layer 1, don't know input structure yet
        if self.num_layers == 1:
            if input_size == None:
                print('in first layer, you need to specity the input size.')
                return
            self.num_neurons[self.num_layers] = num_neurons
            if initialization == 'normal':
                self.weights[self.num_layers] = np.random.normal(scale=1, 
                                                   size=(num_neurons, input_size))
            elif initialization == 'scaled_normal':
                self.weights[self.num_layers] = (np.random.normal(scale=1, 
                                                    size=(num_neurons, input_size))
                                                   /self.num_neurons[1])
            elif initialization == 'uniform_random':
                self.weights[self.num_layers] = np.random.random(size=(num_neurons, input_size))
            
            #self.weights[self.num_layers] = None
            self.biases[self.num_layers] = np.zeros(shape=(num_neurons,1))
            print('add layer ', self.num_layers)
            self.num_layers += 1
            
        else:
            if initialization == 'normal':
                self.weights[self.num_layers] = np.random.normal(scale=1, 
                                                                 size=(num_neurons, self.num_neurons[self.num_layers-1]))
            elif initialization == 'scaled_normal':
                self.weights[self.num_layers] = (np.random.normal(scale=1, 
                                                                  size=(num_neurons, self.num_neurons[self.num_layers-1]))
                                                   /self.num_neurons[1])
            elif initialization == 'uniform_random':
                self.weights[self.num_layers] = np.random.random(size=(num_neurons, self.num_neurons[self.num_layers-1]))
            self.biases[self.num_layers] = np.zeros(shape=(num_neurons, 1))
            self.num_neurons[self.num_layers] = num_neurons
            print('add layer ', self.num_layers)
            self.num_layers += 1
            
    # this will overflow
#     def sigmoid(self, x, derivative=False):
#       print(x)
#         if not derivative:
#             return 1.0/(1.0+np.exp(-x))
#         else:
#             return self.sigmoid(x)*(1-self.sigmoid(x))

    def tanh(self, x, derivative=False):
        pass
    
    def sigmoid(self, x, derivative=False):
        if not derivative:
            return expit(x)
        else:
            return (1-expit(x))*expit(x)
    
    # leaky relu
    def relu(self, x, derivative=False, leaky=False, alpha=0.01):
        result = []
        if not derivative:
            if leaky:
                return np.array([alpha*a if a < 0 else a for a in x]).reshape(-1,1)
            else:
                return np.array([a*(a>0) for a in x]).reshape(-1,1)
            #return np.apply_along_axis(lambda a: alpha*a if a < 0 else a, 1, x).reshape(-1,1)
        else:
            if leaky:
                return np.array([alpha if a < 0 else 1 for a in x]).reshape(-1,1)
            else:
                return np.array([0 if a < 0 else 1 for a in x]).reshape(-1,1)
            #return np.apply_along_axis(lambda a: alpha if a < 0 else 1, 1, x).reshape(-1,1)
    
    def predict(self, x):
        if not self.fitted:
            print('not fitted')
            return
        # if we have 3 weights, they are w_1, w_2, w_3
        # which means we have 3-layers (exclude input layer)
        a = np.ndarray.astype(x, np.float32).reshape((len(x), 1))
        ##print('input: \n', a)
        for i in range(1, self.num_layers+1):
            W = self.weights[i]
            #print('w: \n ', W)
            b = self.biases[i]
            #print('b: \n', b)
            z = np.dot(W,a) + b
            #print('z: \n', z)
            a = self.f(z)
            #print('a: \n', a)
        return a
    
    def fit(self, training_data, batch_size=64, epochs=10, activation_function='sigmoid', 
            learning_rate=0.001, decreasing_rate=0.9, test_data=None, verbose=False): 
        
        self.f = self.activation_functions[activation_function]
        # initialize w_1 
#         if not self.fitted:
#             num_attributes = training_data[0][0].shape[0]
            
#             self.weights[1] = np.random.normal(scale=0.5, 
#                             size=(self.num_neurons[1], num_attributes))
        if self.auto and not self.fitted:
            print("did not specify the output layer, but set 'auto', automatically added output layer based on the shape of the target.")
            print('classes:', len(np.unique(training_data[1])))
            self.add_layer(len(np.unique(training_data[1])))
            self.num_layers -= 1
        self.fitted = True
        print('start training!')
        if type(training_data[0]) == pd.DataFrame:
            x = training_data[0].copy().values
        else:
            x = training_data[0].copy()
        if type(training_data[1]) == pd.DataFrame:
            y = training_data[1].copy().values
        else:
            y = training_data[1].copy()
            
        for i in range(epochs):
            random_index = list(range(training_data[0].shape[1]))
            #print('random_index:', random_index)
            #print(x)
            random.shuffle(random_index)
            #print(self.weights[2])
            #random.shuffle(training_data)
            #print(type(x), type(y))
            batches = [(x[:, k:k+batch_size], y[:, k:k+batch_size]) for k in range(0, training_data[0].shape[1], batch_size)]
            self.seesee['batches'] = batches
            #print('nb_batches:', len(batches))
            for batch, j in zip(batches, range(len(batches))):
                #print(len(batch))
            
                if verbose and j % 20 == 0:
                    print('batch {} / {}'.format(i, len(batches)))
                self.update_mini_batch(batch, learning_rate, batch_size=batch_size)
            
            if test_data:
                print('Epoch {}: {} / {}'.format(i, self.evaluate(test_data), len(test_data[0])))
            else:
                print('Epoch {} complete'.format(i))
            learning_rate *= decreasing_rate
            #print(self.weight_adjustments)
    # x: inputs column by column
    # y: targets column by column
    def backpropagation(self, x, y):   
        bias_gradients = {}
        weight_gradients = {}
        # 1. feedforward
        activations = {0:x.reshape(len(x), 1)} # save all 'a' layer by layer
        #print('x_shape:', x.shape)
        activation = activations[0]
        self.seesee['activations'] = []
        self.seesee['activations'].append(activation)
        zs = {} # save all 'z'(inputs) layer by layer
        for i in range(1, self.num_layers+1):
            w = self.weights[i]
            b = self.biases[i]
            activation = activation.reshape((w.shape[1], len(activation)//w.shape[1]))
            #print(i, 'w_shape:', w.shape, '  b_shape:', b.shape)
            #print(i, 'activation_shape:', activation.shape)
            z = np.dot(w, activation) + b
            #print(i, 'z_shape:', z.shape)
            zs[i] = z
            activation = self.f(z)
            self.seesee['activations'].append(activation)
            activations[i] = activation

        # 2. backpropagation
        # 2.1 compute delta of output layer first
        deltas = {}
        delta = None                                 
        deltas[self.num_layers-1] = 0
        # 2.2 compute delta of each layer
        for i in range(self.num_layers, 0, -1):
            z = zs[i]
            activation = activations[i]
            # output layer
            if i == self.num_layers:
                print('diff:', activation - y) 
                print()
                print('z:', z)
                print()
                print('dz:', self.f(z,derivative=True))
                delta = (activation - y) * self.f(z,derivative=True)
                deltas[i] = delta
            else:
                da_dz = self.f(z, derivative=True)
                delta = deltas[i+1] # delta l+1
                #print(self.weights[i+1].T.shape, delta.shape)
                delta = np.dot(self.weights[i+1].T, delta) * da_dz # delta l
                deltas[i] = delta
        self.seesee['deltas'] = deltas
        # 2.3 compute delta_(l) * a_(l-1).T
        for i in range(self.num_layers, 0, -1):
            #print(deltas[i].shape, activations[i-1].T.shape)
            weight_gradients[i] = np.dot(deltas[i], activations[i-1].T)
            bias_gradients[i] = deltas[i]
        return weight_gradients, bias_gradients
            
    def update_mini_batch(self, batch_data, learning_rate, batch_size=None):
        weight_adjustments = {}
        bias_adjustments = {}
        #print(batch_data)
        _x = batch_data[0]
        _y = batch_data[1]
        #print('x_shape:', _x.shape, 'y_shape:', _y.shape)
        self.count += 1
        for i in range(batch_size):
            # dC/dw and dC/db of each layer
            x = batch_data[0][:, i]
            y = batch_data[1][:, i]
            weight_gradients, bias_gradients = self.backpropagation(x, y)
            #print(weight_gradients, bias_gradients)
            # one batch update once, sum up all gradients first
            for j in range(1, self.num_layers+1):
                weight_adjustments[j] = np.ndarray.astype(weight_adjustments.get(j, np.zeros(shape=weight_gradients[j].shape)), np.float32) + np.ndarray.astype(weight_gradients[j], np.float32)
                bias_adjustments[j] = np.ndarray.astype(bias_adjustments.get(j, np.zeros(shape=bias_gradients[j].shape)), np.float32) + np.ndarray.astype(bias_gradients[j], np.float32)
                #print(weight_adjustments[j])
        self.weight_adjustments = weight_adjustments
        # update with mean of gradients
        for i in range(self.num_layers, 0, -1):
            self.weights[i] -= learning_rate*weight_adjustments[i]/batch_size
            self.biases[i] -= learning_rate*bias_adjustments[i]/batch_size
        
    def evaluate(self, test_data):
        if type(test_data[0]) == pd.DataFrame:
            _x = test_data[0].copy().values
        else:
            _x = test_data[0].copy()
        if type(test_data[1]) == pd.DataFrame:
            _y = test_data[1].copy().values
        else:
            _y = test_data[1].copy()
        
        
        results = []
        for i in range(test_data[0].shape[1]):
            
            x = _x[:, i]
            y = _y[:, i]
            #print(x.shape, y)
            results.append((np.argmax(self.predict(x)), y[0]))
        #print(results)
        #results = [(np.argmax(self.predict(x)), y) for (x, y) in test_data]
        return sum(int(x == y) for (x, y) in results)
    
```
